```python
from sklearn import linear_model
import numpy as np
```

Load the data

```python
import pandas as pd

train = pd.read_csv('train.csv')

test = pd.read_csv('test.csv')
```

Keep the data of ID.

```python
predictions = pd.DataFrame(test['ID']).iloc[0:len(test)]
```

```python
# Create linear regression object
regr = linear_model.LinearRegression()
```

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
```

Split the data of train into train_x and train_y.

```python
train_y = pd.DataFrame(train['target'])
```

```python
train_x = train

del train_x['ID']
del train_x['target']
```

```python
test_x = test
del test_x['ID']
```

Delete the column with null values.

```python
dellist=[]
for i in train_x:
    if(train_x[i].isnull().sum()/len(train_x) != 0):
        print(i,train_x[i].isnull().sum()/len(train_x))
        dellist.append(i)
```

```python
for i in dellist:
    del train_x[i]
    del test_x[i]
```

Use label encoding to transfer categorial variables.

```python
train_x_le1 = train_x.fillna(train_x.mean()).apply(le.fit_transform)
```

```python
regr.fit(train_x_le1, train_y)
```

```python
# Evaluation metrics on test data
# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
 % np.mean((regr.predict(train_x_le1) - train_y) ** 2))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(train_x_le1, train_y))
```

```python
test_x_le1 = test_x.apply(le.fit_transform)
```

```python
val_le1 = regr.predict(test_x_le1)
```

```python
list=[]
for i in val_le1:
    if(abs(float(i))>1):
        list.append(1)
    elif(abs(float(i))<0):
        list.append(0)
    else:
        list.append(abs(float(i)))
```

```python
val_le1=pd.Series(list)
```

```python
predictions['target']=val_le1
```

```python
predictions.to_csv('predictions.csv',index = False)
```
